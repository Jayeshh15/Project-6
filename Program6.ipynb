```python
class FileManager:
    def __init__(self, filename="journal.txt"):
        self.filename = filename
```

```python
class AddEntry(FileManager):
    def add_entry(self):
        try:
            text = input("Enter your journal entry: ")
            with open(self.filename, "a") as file:
                file.write(text + "\n")
            print("Entry added successfully!")
        except Exception as e:
            print("Error:", e)
```

```python
class ViewEntries(FileManager):
    def view_entries(self):
        try:
            with open(self.filename, "r") as file:
                content = file.read()
                if content.strip() == "":
                    print("No entries found.")
                else:
                    print("\n--- Journal Entries ---")
                    print(content)
        except FileNotFoundError:
            print("Error: Journal file does not exist. Please add entry first.")
```

```python
class SearchEntry(FileManager):
    def search_entry(self):
        keyword = input("Enter keyword to search: ")
        try:
            with open(self.filename, "r") as file:
                lines = file.readlines()
                found = False
                print("\nMatching Entries:")
                for line in lines:
                    if keyword.lower() in line.lower():
                        print(line.strip())
                        found = True
                if not found:
                    print("No record found for the keyword.")
        except FileNotFoundError:
            print("Error: Journal file does not exist.")
```

```python
class DeleteEntry(FileManager):
    def delete_all(self):
        confirm = input("Are you sure you want to delete all entries? (yes/no): ")
        if confirm.lower() == "yes":
            try:
                open(self.filename, "w").close()
                print("All entries deleted successfully!")
            except Exception as e:
                print("Error:", e)
        else:
            print("Operation cancelled.")
```

```python
class JournalApp(AddEntry, ViewEntries, SearchEntry, DeleteEntry):
    def menu(self):
        while True:
            print("\nMenu:")
            print("1. Add New Entry")
            print("2. View Entries")
            print("3. Search Entry")
            print("4. Delete All Entries")
            print("5. Exit")

            choice = input("Enter your choice: ")

            if choice == "1":
                self.add_entry()

            elif choice == "2":
                self.view_entries()

            elif choice == "3":
                self.search_entry()

            elif choice == "4":
                self.delete_all()

            elif choice == "5":
                print("Thanks for using Personal Journal Manager.")
                break

            else:
                print("Invalid option. Please select a valid option.")
```

```python
app = JournalApp()
app.menu()
```
